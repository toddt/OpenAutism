```python
# Histograms for frequency of outliers for thresholds 0.2 and 0.4

from scipy.stats import mode
pathway = '/om/user/rezzo/'

Strict = []
Lenient = []

with open(pathway+"INCLUDE_final.csv", "r") as infile:
    X = [list(map(str, line.split())) for line in infile]
    for line in X:
        Strict.append(float(line[4]))
        Lenient.append(float(line[6]))

cont = ['S', 'L']  

for thresh in cont:        
    if thresh is "S":
        fig, ax = plt.subplots(figsize=(8, 5))
        MAX = max(Strict)
        Mode = mode(Strict)
        Median = np.median(Strict)
        Y = range(0,int(MAX))
        Long = len(Strict)
        counts, bins, bars = plt.hist(Strict, bins=Y, normed=True)
        title = 'Frequency_Outliers_0.2'
        ax.set_xlabel('Frequency of Outliers per Run (FDs > 0.2 mm)')
        ax.set_title('Frequency of Outliers for Thresh 0.2')
        ax.annotate("Mode = "+str(Mode[0][0]), xy=(0, 0), xytext=(112, 0.16))
        ax.annotate("Median = "+str(Median), xy=(0, 0), xytext=(112, 0.15))
        ax.annotate("Total Runs = "+str(Long), xy=(0, 0), xytext=(112, 0.14))
        #print(Y)
    elif thresh is "L":
        fig, ax = plt.subplots(figsize=(8, 5))
        #MAX = max(Lenient)
        Mode = mode(Lenient)
        Median = np.median(Lenient)
        #Y = range(0,int(MAX))    
        Long = len(Lenient)
        counts, bins, bars = plt.hist(Lenient, bins=Y, normed=True) 
        title = 'Frequency_Outliers_0.4'
        ax.set_xlabel('Frequency of Outliers per Run (FDs > 0.4 mm)')
        ax.set_title('Frequency of Outliers for Thresh 0.4')
        ax.annotate("Mode = "+str(Mode[0][0]), xy=(0, 0), xytext=(112, 0.16))
        ax.annotate("Median = "+str(Median), xy=(0, 0), xytext=(112, 0.15))
        ax.annotate("Total Runs = "+str(Long), xy=(0, 0), xytext=(112, 0.14))
        #print(Y)
    ax.grid(False)
    ax.set_ylabel('% Runs')

    plt.ylim((0, .17)) 
    plt.xlim((0, MAX)) 

if thresh is "S":    
    plt.savefig('./Histograms_Motion/'+title+'.png')
else:
    plt.savefig('./Histograms_Motion/'+title+'.png')
```
